```python
import gradio as gr

SKILLS = {
    "python": "✅ Data Science → Machine Learning → AI Engineering",
    "web": "✅ HTML/CSS → JavaScript → React → Node.js",
    "ai": "✅ Deep Learning → NLP → Computer Vision → GenAI",
    "data": "✅ Pandas → SQL → Statistics → Visualization",
    "cloud": "✅ AWS → Docker → Kubernetes → DevOps",
}

def recommend(skill):
    result = SKILLS.get(skill.lower().strip())
    return result if result else "❌ Try: python, web, ai, data, cloud"

gr.Interface(
    fn=recommend,
    inputs=gr.Textbox(label="Enter Skill", placeholder="e.g. python"),
    outputs=gr.Textbox(label="Your Roadmap"),
    title="🚀 Skill Recommender",
    examples=[["python"], ["web"], ["ai"], ["data"], ["cloud"]]
).launch(share=True)
```

```python
import gradio as gr

SKILLS = {
    "python": {
        "emoji": "🐍",
        "roadmap": ["Basics & Syntax", "OOP Concepts", "Libraries (NumPy, Pandas)", "Machine Learning", "AI Engineering"],
        "level": "Beginner → Advanced",
        "tip": "Start with 30 mins daily on Python basics!"
    },
    "web": {
        "emoji": "🌐",
        "roadmap": ["HTML & CSS", "JavaScript", "React / Vue", "Node.js", "REST APIs"],
        "level": "Beginner → Intermediate",
        "tip": "Build small projects like a portfolio site early on!"
    },
    "ai": {
        "emoji": "🤖",
        "roadmap": ["Math & Stats", "Deep Learning", "NLP", "Computer Vision", "Generative AI"],
        "level": "Intermediate → Advanced",
        "tip": "Take Andrew Ng's course on Coursera first!"
    },
    "data": {
        "emoji": "📊",
        "roadmap": ["Excel & SQL", "Pandas & NumPy", "Data Visualization", "Statistics", "Dashboards"],
        "level": "Beginner → Intermediate",
        "tip": "Practice on real datasets from Kaggle!"
    },
    "cloud": {
        "emoji": "☁️",
        "roadmap": ["Linux Basics", "Networking", "AWS / GCP / Azure", "Docker", "Kubernetes"],
        "level": "Intermediate → Advanced",
        "tip": "Get one cloud certification to boost your resume!"
    },
    "cybersecurity": {
        "emoji": "🔒",
        "roadmap": ["Networking Basics", "Linux", "Ethical Hacking", "Cryptography", "Pen Testing"],
        "level": "Beginner → Expert",
        "tip": "Try TryHackMe for hands-on practice!"
    },
}

def recommend(skill):
    key = skill.strip().lower()
    match = next((k for k in SKILLS if key in k or k.startswith(key)), None)

    if not match:
        keys = ", ".join(SKILLS.keys())
        return f"❌ Skill not found!\n\n💡 Available skills: {keys}"

    info = SKILLS[match]
    steps = "\n".join(f"  Step {i+1} → {t}" for i, t in enumerate(info["roadmap"]))

    return f"""
🎯 SKILL     : {info['emoji']} {match.title()}
📈 LEVEL     : {info['level']}

🗺️ YOUR ROADMAP:
{steps}

💡 PRO TIP   : {info['tip']}

🚀 Start today and track your progress daily!
""".strip()

# ── UI ──────────────────────────────────────────────────────────────
with gr.Blocks(theme=gr.themes.Soft(), title="Skill Recommender") as app:

    gr.Markdown("# 🚀 Skill Path Recommender")
    gr.Markdown("### Find your personalized learning roadmap in seconds!")
    gr.Markdown("---")

    with gr.Row():
        skill_input = gr.Textbox(
            label="🔍 Enter a Skill",
            placeholder="Type: python, web, ai, data, cloud, cybersecurity",
            scale=4
        )
        submit_btn = gr.Button("Get Roadmap 🎯", variant="primary", scale=1)

    # Quick pick buttons
    gr.Markdown("**⚡ Quick Pick:**")
    with gr.Row():
        for skill, info in SKILLS.items():
            btn = gr.Button(f"{info['emoji']} {skill.title()}", size="sm")
            btn.click(fn=lambda s=skill: s, outputs=skill_input)

    gr.Markdown("---")
    output = gr.Textbox(label="📋 Your Roadmap", lines=10, interactive=False)

    gr.Markdown("---")
    gr.Markdown("<center>Made with ❤️ using Gradio</center>")

    # Events
    submit_btn.click(fn=recommend, inputs=skill_input, outputs=output)
    skill_input.submit(fn=recommend, inputs=skill_input, outputs=output)

app.launch(share=True)
```
